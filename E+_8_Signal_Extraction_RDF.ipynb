## Some useful functions

```python
def update_legend_position(legend):
    """
    Dynamically adjust the legend's size based on the number of entries.
    """
    # Calculate the number of entries and height
    n_entries = legend.GetNRows()
    legend_height = 0.04 * n_entries  # Adjust this scaling factor if needed

    # Fixed horizontal and vertical coordinates
    x1, x2 = 0.2, 0.4
    y2 = 0.9
    y1 = y2 - legend_height  # Adjust the lower boundary based on the height

    # Update legend position
    legend.SetX1NDC(x1)
    legend.SetX2NDC(x2)
    legend.SetY1NDC(y1)
    legend.SetY2NDC(y2)
    legend.Draw()
```

```python
def adjust_histogram_range(histogram):
    """
    Adjust the x-axis range of a ROOT histogram to include only the bins with non-zero content.

    This approach is used because manually calculating the min/max values with methods like
    `data.Min/Max("variable").GetValue()` can be computationally expensive for large datasets.
    Instead, we let ROOT automatically determine the range of the histogram using (0, 0) as
    the axis limits. Then, we dynamically adjust the range based on the non-empty bins of
    the histogram to ensure the x-axis is restricted to meaningful data without displaying
    unnecessary empty space.

    Parameters:
        histogram (ROOT.TH1): The histogram whose range needs adjustment.

    Returns:
        tuple: A tuple containing the new min and max values of the x-axis.
    """


    # Fill the histogram (this step is necessary for ROOT to compute its bin contents)
    histogram.GetValue()

    # Access the x-axis of the histogram for further operations
    x_axis = histogram.GetXaxis()

    # Find the first and last bins that contain non-zero data
    # `FindFirstBinAbove(0)` locates the first bin with a value greater than 0 (first non-empty bin)
    # `FindLastBinAbove(0)` locates the last bin with a value greater than 0 (last non-empty bin)
    min_bin = histogram.FindFirstBinAbove(0)  # Index of the first non-empty bin
    max_bin = histogram.FindLastBinAbove(0)   # Index of the last non-empty bin

    # Retrieve the edges of the first and last bins to define the meaningful range of data
    # `GetBinLowEdge(min_bin)` gives the lower edge of the first non-empty bin
    # `GetBinUpEdge(max_bin)` gives the upper edge of the last non-empty bin
    min_value = x_axis.GetBinLowEdge(min_bin)  # Lower bound of the meaningful data range
    max_value = x_axis.GetBinUpEdge(max_bin)   # Upper bound of the meaningful data range

    # Adjust the x-axis range to only include the bins with non-zero data
    # This ensures the histogram does not display unnecessary empty space
    x_axis.SetRangeUser(min_value, max_value)

    # Return the new range for logging or further use
    return min_value, max_value
```

## Johnson PDF




The Johnson PDF is a versatile probability density function designed to describe distributions that deviate from the symmetric, bell-shaped Gaussian.
It is particularly well-suited for modeling data with asymmetry, heavy tails, or other irregularities.

**Parameters:**

- **mean**: Represents the central value or the peak of the distribution.
- **sigma**: Determines the width or spread of the distribution around the mean.

- **gamma**: A skewness parameter that adjusts the asymmetry of the distribution.
  - Typical range: `[-1.0, 1.0]`.
  - Negative values result in left-skewed distributions, while positive values create right-skewed ones.

- **delta**: A shape parameter that influences the peakedness and tail behavior of the distribution.
  - Typical range: `[0.1, 5.0]`.
  - Smaller values produce wider distributions with heavier tails, while larger values create sharper peaks with lighter tails.

**When to Use the Johnson PDF?**

- The Johnson PDF is ideal for situations where the signal distribution cannot be adequately modeled by a Gaussian due to asymmetry or unusual tail behavior.
- It is commonly applied in high-energy physics and other fields requiring detailed modeling of complex data features.

**Additional Resources:**

- For more detailed information, refer to the [RooJohnson documentation](https://root.cern.ch/doc/master/classRooJohnson.html).
- To use this PDF in ROOT, it can be invoked as `RooJohnson`.


```python
from ROOT import TMath

mass_J = ROOT.RooRealVar("mass_J", "candidate_vMass", min_value_new, max_value_new)
mass_J.setBins(bins)

fit_frame_J = mass_J.frame(Title="Johnson PDF with Pull Distribution")

data_hist_J = ROOT.RooDataHist(
    "data_hist_J", "Data Histogram J",
    ROOT.RooArgList(mass_J), ROOT.RooFit.Import(hist_data6_mass.GetValue())
)


total_yield_J = data_hist_J.sumEntries()

# Johnson Signal Model
mean_J = ROOT.RooRealVar("mean", "mean of gaussian J", 5.37, 5.3, 5.4)
sigma_J = ROOT.RooRealVar("sigma", "width J", 0.012, 0.01, 0.015)
gamma_J = ROOT.RooRealVar("gamma", "gamma J", 0., -1., 1.)
delta_J = ROOT.RooRealVar("delta", "delta J", 1.0, 0.1, 2.0)
signal_J = ROOT.RooJohnson("signal", "signal J", mass_J, mean_J, sigma_J, gamma_J, delta_J)


# Exponential Background Model
c0_J = ROOT.RooRealVar("c0", "c0 J", -0.2, -2, -0.1)
background_J = ROOT.RooExponential("background", "Background J", mass_J, c0_J)

# Yield Parameters
Nsig_J = ROOT.RooRealVar("Nsig", "Johnson signal yield J", 0.5 * total_yield_J, 0, total_yield_J)
Nbkg_J = ROOT.RooRealVar("Nbkg", "Background yield J", 0.5 * total_yield_J, 0, total_yield_J)

# Combine Signal and Background Models
model_J = ROOT.RooAddPdf(
    "model_J", "Signal + Background J",
    ROOT.RooArgList(signal_J, background_J),
    ROOT.RooArgList(Nsig_J, Nbkg_J)
)

# Fit Model to Data
fit_result_J = model_J.fitTo(data_hist_J, ROOT.RooFit.Save())


mean_J_val = mean_J.getVal()
sigma_J_val = sigma_J.getVal()

# 3-sigma window
minus_3s_J = mean_J_val - 3.0 * sigma_J_val
plus_3s_J = mean_J_val + 3.0 * sigma_J_val
print(f"3σ interval: {minus_3s_J:.5f} to {plus_3s_J:.5f}")

# 2-sigma window
minus_2s_J = mean_J_val - 2.0 * sigma_J_val
plus_2s_J = mean_J_val + 2.0 * sigma_J_val
print(f"2σ interval: {minus_2s_J:.5f} to {plus_2s_J:.5f}")


# Define 3σ and 2σ signal windows
mass_J.setRange("range_3s", minus_3s_J, plus_3s_J)
mass_J.setRange("range_2s", minus_2s_J, plus_2s_J)

# --- Signal yield ---
frac_signal_3s = signal_J.createIntegral(ROOT.RooArgSet(mass_J), ROOT.RooFit.NormSet(mass_J), ROOT.RooFit.Range("range_3s"))
S_3s = frac_signal_3s.getVal() * Nsig_J.getVal()

frac_signal_2s = signal_J.createIntegral(ROOT.RooArgSet(mass_J), ROOT.RooFit.NormSet(mass_J), ROOT.RooFit.Range("range_2s"))
S_2s = frac_signal_2s.getVal() * Nsig_J.getVal()

# --- Background yield ---
frac_bkg_3s = background_J.createIntegral(ROOT.RooArgSet(mass_J), ROOT.RooFit.NormSet(mass_J), ROOT.RooFit.Range("range_3s"))
B_3s = frac_bkg_3s.getVal() * Nbkg_J.getVal()

frac_bkg_2s = background_J.createIntegral(ROOT.RooArgSet(mass_J), ROOT.RooFit.NormSet(mass_J), ROOT.RooFit.Range("range_2s"))
B_2s = frac_bkg_2s.getVal() * Nbkg_J.getVal()

# --- Significance in 3σ window ---
T_3s = S_3s + B_3s
significance_3s = S_3s / TMath.Sqrt(T_3s)

# --- Purity in 2σ window ---
T_2s = S_2s + B_2s
purity_2s = S_2s / T_2s


print(f"S (±3σ): {S_3s:.1f}")
print(f"B (±3σ): {B_3s:.1f}")
print(f"Significance (±3σ): {significance_3s:.3f}")

print(f"S (±2σ): {S_2s:.1f}")
print(f"B (±2σ): {B_2s:.1f}")
print(f"Purity (±2σ): {purity_2s:.4f}")


data_hist_J.plotOn(fit_frame_J)
model_J.plotOn(fit_frame_J)
model_J.plotOn(fit_frame_J, ROOT.RooFit.Components("signal"), ROOT.RooFit.LineColor(ROOT.kRed))
model_J.plotOn(fit_frame_J, ROOT.RooFit.Components("background"), ROOT.RooFit.LineColor(ROOT.kGreen))
model_J.paramOn(fit_frame_J, ROOT.RooFit.Layout(0.582, 0.9, 0.9))
model_J.plotOn(fit_frame_J)



chi2_J = model_J.createChi2(data_hist_J)
npar_J = fit_result_J.floatParsFinal().getSize()
ndof_J = bins - npar_J
chi2_ndf_J = chi2_J.getVal() / ndof_J

hpull_J = fit_frame_J.pullHist()
pull_frame_J = mass_J.frame(Title=" ")
pull_frame_J.addPlotable(hpull_J, "P")

pull_frame_J.GetXaxis().SetLabelSize(0.07)
pull_frame_J.GetYaxis().SetLabelSize(0.07)
pull_frame_J.GetYaxis().SetTitle("#pm#chi^{2}")
pull_frame_J.GetYaxis().SetTitleSize(0.07)
pull_frame_J.GetYaxis().SetTitleOffset(0.3)
pull_frame_J.GetYaxis().SetRangeUser(-5, +5)


canvas_fit_J = ROOT.TCanvas("canvas_fit_J", "Johnson Fit Canvas with Pulls", 800, 800)
pad1_J = ROOT.TPad("pad1_J", "Pad with fit", 0, 0.3, 1, 1.0)
pad2_J = ROOT.TPad("pad2_J", "Pad with pulls", 0, 0.05, 1, 0.3)
pad1_J.Draw()
pad2_J.Draw()

pad1_J.cd()
fit_frame_J.Draw()

pad2_J.cd()
pull_frame_J.Draw()



line0_J = ROOT.TLine(min_value_new, 0, max_value_new, 0)
lineP_J = ROOT.TLine(min_value_new, 3, max_value_new, 3)
lineM_J = ROOT.TLine(min_value_new, -3, max_value_new, -3)
for line in [line0_J, lineP_J, lineM_J]:
    line.SetLineStyle(2)
    line.SetLineColor(ROOT.kGray + 1)
    line.Draw("same")

latex_chi2_J = ROOT.TLatex(0.15, 0.85, f"#chi^{{2}}/ndf = {chi2_ndf_J:.2f}")
latex_chi2_J.SetNDC()
latex_chi2_J.SetTextSize(0.03)
pad1_J.cd()


# Add vertical lines for ±2σ and ±3σ boundaries (after computing mean_J_val and sigma_J_val)

line_minus_3s = ROOT.TLine(minus_3s_J, 0, minus_3s_J, fit_frame_J.GetMaximum())
line_plus_3s  = ROOT.TLine(plus_3s_J, 0, plus_3s_J, fit_frame_J.GetMaximum())

line_minus_2s = ROOT.TLine(minus_2s_J, 0, minus_2s_J, fit_frame_J.GetMaximum() * 0.9)
line_plus_2s  = ROOT.TLine(plus_2s_J, 0, plus_2s_J, fit_frame_J.GetMaximum() * 0.9)

for line in [line_minus_3s, line_plus_3s, line_minus_2s, line_plus_2s]:
    line.SetLineColor(ROOT.kRed + 1)
    line.SetLineStyle(2)
    line.SetLineWidth(1)
    pad1_J.cd()
    line.Draw("same")



latex_chi2_J.Draw()


canvas_fit_J.Draw()
canvas_fit_J.SaveAs("johnson_fit_with_background_J.png")

```

```python
# Double Gaussian Signal + Exponential Background

mass_DG = ROOT.RooRealVar("mass_DG", "candidate_vMass", min_value_new, max_value_new)
mass_DG.setBins(bins)

data_hist_DG = ROOT.RooDataHist(
    "data_hist_DG", "Data Histogram DG",
    ROOT.RooArgList(mass_DG), ROOT.RooFit.Import(hist_data6_mass.GetValue(), False)
)

total_yield_DG = data_hist_DG.sumEntries()

gauss_mean_DG = ROOT.RooRealVar("#mu1_{DG}", "mean of first gaussian DG", 5.37, 5.35, 5.39)

gauss_sigma1_DG = ROOT.RooRealVar("#sigma1_{DG}", "width of first gaussian DG", 0.012, 0.005, 0.015)
gauss_sigma2_DG = ROOT.RooRealVar("#sigma2_{DG}", "width of second gaussian DG", 0.18, 0.016, 0.20)

# Gaussian models
signal_gauss1_DG = ROOT.RooGaussian("signal_gauss1_DG", "First Gaussian Signal DG", mass_DG, gauss_mean_DG, gauss_sigma1_DG)
signal_gauss2_DG = ROOT.RooGaussian("signal_gauss2_DG", "Second Gaussian Signal DG", mass_DG, gauss_mean_DG, gauss_sigma2_DG)

# Gaussian fraction
frac_gauss_DG = ROOT.RooRealVar("frac_gauss_DG", "fraction of first Gaussian DG", 0.5, 0.3, 0.7)

# Combine Gaussian model
signal_model_DG = ROOT.RooAddPdf(
    "signal_model_DG", "Combined Gaussian Signal DG",
    ROOT.RooArgList(signal_gauss1_DG, signal_gauss2_DG),
    ROOT.RooArgList(frac_gauss_DG)
)

# Exponential Background Model
c0_background_DG = ROOT.RooRealVar("c0_{DG}", "Exponential coefficient DG", -1, -10, 0)
background_model_DG = ROOT.RooExponential("background_DG", "Exponential Background DG", mass_DG, c0_background_DG)

# Yield Parameters
Nsig_DG = ROOT.RooRealVar("N_{sig}^{DG}", "Total signal yield DG", 0.5 * total_yield_DG, 0, total_yield_DG)
Nbkg_DG = ROOT.RooRealVar("N_{bkg}^{DG}", "Background yield DG", 0.5 * total_yield_DG, 0, total_yield_DG)

# Combine Signal and Background Models
model_DG = ROOT.RooAddPdf(
    "model_DG", "Signal + Background DG",
    ROOT.RooArgList(signal_model_DG, background_model_DG), ROOT.RooArgList(Nsig_DG, Nbkg_DG)
)


fit_result_DG = model_DG.fitTo(data_hist_DG, ROOT.RooFit.Save())

chi2_DG = model_DG.createChi2(data_hist_DG)
npar_DG = fit_result_DG.floatParsFinal().getSize()
ndof_DG = bins - npar_DG
chi2_ndf_DG = chi2_DG.getVal() / ndof_DG

fit_frame_DG = mass_DG.frame(Title="Double Gaussian + Exponential Fit DG")
data_hist_DG.plotOn(fit_frame_DG)
model_DG.plotOn(fit_frame_DG)
#model_DG.plotOn(fit_frame_DG, ROOT.RooFit.Components("signal_model_DG"), ROOT.RooFit.LineColor(ROOT.kRed))

model_DG.plotOn(fit_frame_DG, ROOT.RooFit.Components("signal_gauss1_DG"), ROOT.RooFit.LineStyle(ROOT.kDashed), ROOT.RooFit.LineColor(ROOT.kMagenta), ROOT.RooFit.Name("G1"))
model_DG.plotOn(fit_frame_DG, ROOT.RooFit.Components("signal_gauss2_DG"), ROOT.RooFit.LineStyle(ROOT.kDashed), ROOT.RooFit.LineColor(ROOT.kViolet), ROOT.RooFit.Name("G2"))
model_DG.plotOn(fit_frame_DG, ROOT.RooFit.Components("signal_model_DG"), ROOT.RooFit.LineColor(ROOT.kRed), ROOT.RooFit.Name("Signal"))


model_DG.plotOn(fit_frame_DG, ROOT.RooFit.Components("background_DG"), ROOT.RooFit.LineColor(ROOT.kGreen+2), ROOT.RooFit.Name("Background"))
#model_DG.paramOn(fit_frame_DG, ROOT.RooFit.Layout(0.55, 0.6, 0.9))
model_DG.paramOn(
    fit_frame_DG,
    ROOT.RooFit.Parameters(ROOT.RooArgSet(
        gauss_mean_DG,
        gauss_sigma1_DG,
        gauss_sigma2_DG,
        c0_background_DG,
        Nsig_DG,
        Nbkg_DG
    )),
    ROOT.RooFit.Layout(0.55, 0.6, 0.9)
)

model_DG.plotOn(fit_frame_DG)

pull_hist_DG = fit_frame_DG.pullHist()
pull_frame_DG = mass_DG.frame(Title=" ")
pull_frame_DG.addPlotable(pull_hist_DG, "P")

pull_frame_DG.GetXaxis().SetLabelSize(0.07)
pull_frame_DG.GetYaxis().SetLabelSize(0.07)
pull_frame_DG.GetYaxis().SetTitle("#pm#chi^{2}")
pull_frame_DG.GetYaxis().SetTitleSize(0.07)
pull_frame_DG.GetYaxis().SetTitleOffset(0.3)
pull_frame_DG.GetYaxis().SetRangeUser(-5, +5)

canvas_fit_DG = ROOT.TCanvas("canvas_fit_DG", "Double Gaussian + Exponential Fit with Pulls DG", 800, 800)
pad1_DG = ROOT.TPad("pad1_DG", "Pad with fit DG", 0, 0.3, 1, 1.0)
pad2_DG = ROOT.TPad("pad2_DG", "Pad with pulls DG", 0, 0.05, 1, 0.3)
pad1_DG.Draw()
pad2_DG.Draw()

pad1_DG.cd()
fit_frame_DG.Draw()

legend_DG = ROOT.TLegend(0.15, 0.55, 0.45, 0.78)
legend_DG.SetTextSize(0.035)
legend_DG.SetBorderSize(0)
legend_DG.SetFillStyle(0)

legend_DG.AddEntry(fit_frame_DG.findObject("G1"), "1st Gaussian", "l")
legend_DG.AddEntry(fit_frame_DG.findObject("G2"), "2nd Gaussian", "l")
legend_DG.AddEntry(fit_frame_DG.findObject("Signal"), "Signal Model", "l")
legend_DG.AddEntry(fit_frame_DG.findObject("Background"), "Background", "l")
legend_DG.Draw()


pad2_DG.cd()
pull_frame_DG.Draw()

line0_DG = ROOT.TLine(min_value_new, 0, max_value_new, 0)
lineP_DG = ROOT.TLine(min_value_new, 3, max_value_new, 3)
lineM_DG = ROOT.TLine(min_value_new, -3, max_value_new, -3)
for line in [line0_DG, lineP_DG, lineM_DG]:
    line.SetLineStyle(2)
    line.SetLineColor(ROOT.kGray + 1)
    line.Draw("same")

latex_chi2_DG = ROOT.TLatex(0.15, 0.85, f"#chi^{{2}}/ndf = {chi2_ndf_DG:.2f}")
latex_chi2_DG.SetNDC()
latex_chi2_DG.SetTextSize(0.03)
pad1_DG.cd()
latex_chi2_DG.Draw()

canvas_fit_DG.Draw()

canvas_fit_DG.SaveAs("double_gaussian_exponential_fit_DG.png")
```
